```python
import geopandas as gpd
import matplotlib.pyplot as plt
from libpysal.weights import KNN
import esda
import numpy as np

# Load the GeoDataFrame
cat = gpd.read_file("G:/My Drive/INVESTIGACION/PAPERS/ELABORACION/Modelo_SAR/DATA/df_catchments_kmeans.gpkg")

# Variables to analyze
variables = ['area', 'hypso_inte', 'slope_mean', 'rainfallAnnual_mean']
k_values = list(range(1, 11)) + list(range(10, 501, 10))  # K from 1 to 10 by 1, then 10 to 500 by 10

# Data storage for results
results = {var: {'k_values': [], 'mi_values': [], 'significant_percentages': []} for var in variables}

for variable in variables:
    for k in k_values:
        # Create a K-nearest neighbors weights matrix
        w_knn = KNN.from_dataframe(cat, k=k)
        
        # Identify islands (disconnected components)
        islands = [k for k, v in w_knn.neighbors.items() if len(v) == 0]

        # Filter out islands from the GeoDataFrame
        cat_filtered = cat.drop(index=islands)

        # Create a new KNN weights matrix without islands
        w_knn_filtered = KNN.from_dataframe(cat_filtered, k=k)

        # Calculate local Moran's I (using filtered data)
        lisa = esda.Moran_Local(cat_filtered[variable], w_knn_filtered)

        # Calculate the global Moran's I
        mi = esda.Moran(cat_filtered[variable], w_knn_filtered)

        # Calculate percentage of significant observations
        significant = sum(lisa.p_sim < 0.05)
        percentage_significant = (significant / len(lisa.p_sim)) * 100

        # Store results
        results[variable]['k_values'].append(k)
        results[variable]['mi_values'].append(mi.I)
        results[variable]['significant_percentages'].append(percentage_significant)

# Plotting results
for variable in variables:
    plt.figure(figsize=(10, 6))
    k_values = results[variable]['k_values']
    mi_values = results[variable]['mi_values']
    percentages = results[variable]['significant_percentages']

    plt.scatter(k_values, mi_values, c='blue', alpha=0.7, label='Moran\'s I')
    for i, perc in enumerate(percentages):
        plt.text(k_values[i], mi_values[i], f'{int(perc)}%', fontsize=9, ha='right')

    # Set axis labels and ticks
    plt.xlabel('Número de vecinos (K)', fontsize=15)  # Changed x-axis label
    plt.ylabel('Indice de Moran', fontsize=15)  # Increased font size for y-axis label
    plt.xticks(fontsize=12)  # Increased font size of x ticks
    plt.yticks(fontsize=12)  # Increased font size of y ticks
    plt.xlim(0, 500)  # Adjusted x-axis to show 0 to 500
    plt.ylim(0, 1)  # Adjusted y-axis to show 0 to 1
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"G:/My Drive/INVESTIGACION/PAPERS/ELABORACION/Modelo_SAR/FIGURAS/{variable}_moran_vs_k_neighbors.png")
    plt.show()
```
